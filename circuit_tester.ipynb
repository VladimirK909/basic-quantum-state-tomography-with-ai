```python
import qiskit
from qiskit.quantum_info import random_statevector, DensityMatrix
```

```python
qc = qiskit.QuantumCircuit(1)
qc.measure_all()
qc.draw('mpl', style='clifford')
```

```python
qc = qiskit.QuantumCircuit(2)
qc.h(0)
qc.cx(0,1)
qc.measure_all()
qc.draw('mpl', style='clifford')
```

```python
qc = qiskit.QuantumCircuit(3)
qc.h(0)
qc.cx(0, 1)
qc.cx(1, 2)
qc.measure_all()
qc.draw('mpl', style='clifford')
```
